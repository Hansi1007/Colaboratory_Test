```python
 import numpy as np
 import matplotlib.pyplot as plt
```

```python
X_train = np.asarray([1,2,3,4])
Y_train = np.asarray([1,3,5,4])

plt.plot(X_train, Y_train, 'b.')
plt.plot(np.asarray([0,5]), 1.1*np.asarray([0,5]) + 0.5)
```

```python
mean_x = np.sum(X_train) / len(X_train)
mean_y = np.sum(Y_train) / len(Y_train)
print(mean_x)
print(mean_y)
```

```python
var_x = np.sum((X_train - mean_x)**2)/ len(X_train)
var_y = np.sum((Y_train - mean_y)**2)/ len(Y_train)
print(X_train - mean_x)
print("Varianz X:", var_x)
print("Varianz Y:", var_y)
```

```python
print(np.var(X_train))
print(np.var(Y_train))
X = np.stack((X_train, Y_train), axis=0)
print(X)
cov = np.cov(X, bias=True)
print(cov)
#print(np.cov(X_train, Y_train, bias=True))
```

```python
cova_xy = np.sum((X_train - mean_x)*(Y_train - mean_y)) / len(X_train)
print(cova_xy)
```

image.png

```python
alpha = cova_xy / var_x
print(alpha)
```

```python
import seaborn as sns

labs = ['X_train', 'Y_train']

sns.heatmap(cov, annot=True, fmt='g', xticklabels=labs, yticklabels=labs)
plt.show()
```

```python
beta = mean_y - alpha * mean_x
print(beta)
```

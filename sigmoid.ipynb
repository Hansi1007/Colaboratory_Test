```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
sig = 1 / ( 1 + np.exp(-x))

sigd = np.exp(x) / ((1+np.exp(x))**2)


plt.plot(x, sig, linewidth=1, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.plot(x, sigd, linewidth=1, color='c')
plt.xlim(-10, 10)
plt.xlabel("x")
plt.legend(fontsize=14)
plt.grid()
plt.show()
```

```python
t = np.linspace(-10, 10, 100)

plt.axhline(y=0, color="black", linestyle="--")
plt.axhline(y=0.5, color="black", linestyle=":")
plt.axhline(y=1.0, color="black", linestyle="--")
plt.axvline(color="grey")

# mit skalierung faktor s
s = 0.5
sig = 1 / (1 + np.exp(-t*s))
plt.plot(t, sig, linewidth=2, color='r', label=r"$\sigma(t) = \frac{1}{1 + e^{-t*0.5}}$")

s = 1
sig = 1 / (1 + np.exp(-t*s))
plt.plot(t, sig, linewidth=2, color='g', label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")

s = 10
sig = 1 / (1 + np.exp(-t*s))
plt.plot(t, sig, linewidth=2, color='b', label=r"$\sigma(t) = \frac{1}{1 + e^{-t*10}}$")


plt.xlim(-10, 10)
plt.xlabel("t")
plt.legend(fontsize=14)
plt.show()
```

```python
def softmax3(z1, z2, z3):
  return np.exp(z1) / (np.exp(z1) + np.exp(z2) + np.exp(z3))
```

```python
z1_input = 2.33
z2_input = -1.46
z3_input = 0.56

print("output class 1 = ", softmax3(z1_input, z2_input, z3_input) )
print("output class 2 = ", softmax3(z2_input, z1_input, z3_input) )
print("output class 2 = ", softmax3(z3_input, z1_input, z1_input) )
```
